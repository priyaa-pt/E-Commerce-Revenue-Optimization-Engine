## exploratory data analysis(eda)

### univariate analysis (numerical columns)

```python
num_cols = df.select_dtypes(include=['float64','int64']).columns

for col in num_cols:
    print(df[col].describe())
    sns.histplot(df[col], kde=True)
    plt.title(f"Distribution of {col}")
    plt.show()
```

```python
for col in num_cols: #outlier analysis
    sns.boxplot(df[col])
    plt.title(f"Outlier Analysis for {col}")
    plt.show()
```

```python
# Univariate Analysis: Categorical Columns

cat_cols = df.select_dtypes(include=['object']).columns

for col in cat_cols:
    print(f"\n----- {col} -----")
    print(df[col].value_counts().head())  # show top 5 categories
    
    plt.figure(figsize=(8,4))
    sns.countplot(data=df, x=col, order=df[col].value_counts().index[:10])
    plt.title(f"Countplot of {col} (Top 10)")
    plt.xticks(rotation=45)
    plt.show()
```

### Bivariate Analysis

```python
# Correlation Heatmap
plt.figure(figsize=(10,6))
sns.heatmap(df[num_cols].corr(), annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()
```

## Bivariate Analysis â€“ Numeric vs Numeric

```python
import matplotlib.pyplot as plt
import seaborn as sns

num_pairs = [
    ('product_price', 'revenue'),
    ('discount_percent', 'revenue'),
    ('quantity', 'revenue'),
    ('delivery_days', 'customer_rating')
]

plt.figure(figsize=(12,10))

for i, (x, y) in enumerate(num_pairs, 1):
    plt.subplot(2, 2, i)
    sns.scatterplot(data=df, x=x, y=y, alpha=0.4)
    plt.title(f"{x} vs {y}")

plt.tight_layout()
plt.show()
```
